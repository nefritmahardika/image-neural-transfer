```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = hub.load ('https://www.kaggle.com/models/google/arbitrary-image-stylization-v1/frameworks/TensorFlow1/variations/256/versions/2')
```

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
content_image = load_image('/content/monas.jpg')
style_image = load_image('/content/frida.jpg')
```

```python
plt.imshow(np.squeeze(content_image))
plt.show()
```

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(stylized_image))
plt.show()
```
